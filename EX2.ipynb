###### compute the standard deviation of gestation time for first babies and others. Does it look like the spread is the same for the two groups

```python
import get_data
import numpy as np
```

```python
Preg_df = get_data.ReadFemPreg()
```

```python
Preg_df.head()
```

```python
Preg_df=Preg_df[['caseid', 'wksgest','outcome','pregordr','birthord','birthwgt_lb','birthwgt_oz','agepreg','finalwgt']]
```

```python
Preg_df.head()
```

```python
first_bbies=Preg_df[Preg_df['pregordr']==1]

variance = np.nanvar(first_bbies['wksgest'])
std = np.sqrt(variance)
 
#OR 
std= np.nanstd(first_bbies['wksgest'])
std

print("First standard deviation: %f" %std)
```

```python
other_bbies=Preg_df[Preg_df['pregordr']!=1]
variance = np.nanvar(other_bbies['wksgest'])
std_o = np.sqrt(variance)
 
#OR 

std_o= np.nanstd(other_bbies['wksgest'])
std_o

print("other babies standard deviation: %f" %std_o)
```

### Conclusion

###### The difference in standard deviation is not statistical significant
